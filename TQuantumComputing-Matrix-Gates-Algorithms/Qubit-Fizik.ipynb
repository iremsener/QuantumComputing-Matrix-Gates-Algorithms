Klasik bilgisayarlar, temel olarak bitlerden oluşan işlem birimleri kullanırken, kuantum bilgisayarlar qubit adı verilen kuantum bitler üzerinden işlem yapar.
```python
from qiskit import *
```

### Süperpozisyon:
Süperpozisyon, bir kuantum sisteminin aynı anda birden fazla durumu alabilme yeteneğini ifade eder. Klasik fizikte bir nesnenin belirli bir anda sadece belirli bir durumda olabileceğini düşünürüz. Ancak kuantum dünyasında, bir parçacığın durumu belirli bir durum yerine, bu durumların bir kombinasyonu olarak ifade edilebilir. Örneğin, bir qubit aynı anda hem "0" hem de "1" durumunda olabilir. Süperpozisyon, kuantum hesaplamaların ve kuantum paralelizminin temelini oluşturur.

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector

# Bir kubitlik bir devre oluşturun
circuit = QuantumCircuit(1)

# Qubite Hadamard kapısı uygulayın
circuit.h(0)

# Devreyi simüle edin
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()

# Bloch küresi gösterimini oluşturun
plot_bloch_multivector(statevector)
```

### Interferans:
Interferans, kuantum mekaniğinin dalga benzeri özelliklerinin bir sonucu olarak ortaya çıkan bir olgudur. İki veya daha fazla kuantum durumu arasında birleştiğinde veya etkileştiğinde, bu durumların sonuçta birbiriyle etkileşerek belirli bir sonuç üretmesi durumudur. Interferans, klasik fizikte de bilinen bir olgudur; örneğin su dalgalarının birleşip veya etkileşip yükseklikleri artabilir veya azalabilir. Ancak kuantum dünyasında, bu etkileşimler daha karmaşık bir şekilde gerçekleşir ve belirli durumların ortaya çıkmasına neden olur.

Kuantum dünyasındaki süperpozisyon ve interferans, kuantum hesaplamalar ve kuantum iletişim gibi alanlarda büyük bir öneme sahiptir. Bu kavramlar, geleneksel mantıkla düşünüldüğünde bazen karmaşıklık yaratsa da, kuantum sistemlerin potansiyelini anlamak ve kullanmak için temel taşları oluşturur.

```python


# Bir kubitlik bir devre oluşturun
circuit = QuantumCircuit(1)

# Qubite Hadamard kapısı uygulayın (süperpozisyon oluşturuluyor)
circuit.h(0)

# Devreyi simüle edin
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()

# Qubite X kapısı (NOT kapısı) uygulayın (interferans oluşturuluyor)
circuit.x(0)

# Devreyi tekrar simüle edin
job = execute(circuit, simulator)
result = job.result()
statevector_after_x = result.get_statevector()

# Bloch küresi gösterimini oluşturun (karşılaştırma yapmak için)
plot_bloch_multivector([statevector, statevector_after_x], ['Süperpozisyon', 'İnterferans'])
```

### Dolanıklık

Quantum entanglement" (kuantum dolanıklık), kuantum mekaniğindeki temel bir olgudur. Klasik fizikte alışık olduğumuz ilişkilerin ötesinde, bir çift veya daha fazla parçacığın, belirli özellikleri (spin, polarizasyon gibi) birbirleriyle karşılıklı olarak belirlendiğinde, bu parçacıkların dolanık olduğu söylenir. Bu durumda, bir parçacığın durumu diğer parçacığın durumunu etkiler ve bu etkileşim anında gerçekleşir, yani ışık hızından daha hızlıdır. Bu durum, klasik fizikteki "nedensellik ilkesi"ne aykırıdır.

Kuantum dolanıklığı, kuantum mekaniğinin temel özelliklerinden biridir ve Einstein, Podolsky ve Rosen (EPR) paradoksuyla da ilgili bir konudur. Kuantum dolanıklığı, "spooky action at a distance" (uzakta tuhaf etkileşim) olarak adlandırılan olayı ifade eder.

Kuantum programlamada, kuantum dolanıklık, kuantum devrelerinde belirli parçacıkların durumlarını birbirleriyle ilişkilendirme ve etkileştirme yeteneğini ifade eder. Bu, kuantum algoritmalarının bazılarının temelini oluşturur. Özellikle, kuantum dolanıklığı kuantum hesaplamalarında kullanılan kubitlerin etkileşimini anlamak ve optimize etmek için önemlidir.

Kuantum dolanıklığı, kuantum iletişimi, kuantum hesaplamaları ve kuantum şifreleme gibi birçok alanda kullanılır. Kuantum bilgisayarlar ve kuantum iletişimi, dolanık durumları kullanarak belirli problemleri daha hızlı ve etkili bir şekilde çözebilirler. Bu nedenle, kuantum programlamada kuantum dolanıklığı anlamak ve yönetmek büyük bir öneme sahiptir.

### Kuantum Bit Durumu (Qubit state) :
 Kuantum bilgisayarlarında kullanılan temel bir kavramdır. Bir qubit (kuantum bit), klasik bitin aksine sadece "0" veya "1" durumu yerine, aynı anda hem "0" hem de "1" durumunu aynı anda taşıyabilen bir kuantum mekanik birimdir. Dolayısıyla, bir qubitin durumu, klasik bir bitin sadece iki durumu olan durumundan daha karmaşıktır.

Qubitlerin durumu, bir amplitüd ve bir faz açısıyla ifade edilir. Genellikle |0⟩ ve |1⟩ gibi ket notasyonu kullanılarak gösterilir. Örneğin, bir qubit "α|0⟩ + β|1⟩" şeklinde ifade edilebilir, burada α ve β katsayıları qubitin süperpozisyon durumunu belirtir.

Qubit state, bir qubit'in belirli bir anlık durumunu ifade eder. Bu durum, qubitin ölçülmeden önceki kuantum mekaniksel hali olarak düşünülebilir. Qubit state, kuantum hesaplamalarında, kuantum devrelerinde ve algoritmalarında önemli bir rol oynar. Kuantum bilgisayarlar, qubit state'lerinin manipülasyonu ve kontrolü sayesinde belirli hesaplamaları gerçekleştirirler.

Kuantum mekaniğinde bir qubit, karmaşık bir sayı cinsinden temsil edilir. Süperpozisyon durumu, qubitin "0" ve "1" durumlarının bir kombinasyonunu ifade eder. Bu kombinasyonu temsil eden bir qubit :

|ψ⟩ = α|0⟩ + β|1⟩


|ψ⟩, qubit durumunu temsil eden kuantum durum vektörüdür.
α ve β karmaşık sayılardır. α^2 ve β^2 toplamı 1'e eşit olmalıdır (olasılık korunumu ilkesi).
|0⟩ ve |1⟩, sırasıyla "spin up" ve "spin down" durumlarını temsil eden durum vektörleridir.
Örnek olarak, eğer bir qubit süperpozisyonda olsun ve α = 1/√2, β = 1/√2 olsun:

|ψ⟩ = (1/√2)|0⟩ + (1/√2)|1⟩

Bu, qubitin hem "0" hem de "1" durumunda aynı anda bulunduğu bir süperpozisyon durumunu ifade eder.

Bu matematiksel formül, bir qubitin süperpozisyon durumunu temsil etmek için kullanılır. α ve β katsayıları, qubitin "0" ve "1" durumları arasındaki olasılık dağılımını belirler.

```python
# Bir kubitlik bir devre oluşturun
circuit = QuantumCircuit(1)

# Qubite belirli bir durum verin (örneğin |0⟩ durumu)
circuit.initialize([1, 0], 0)

# Devreyi simüle edin
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()

# Bloch küresi gösterimini oluşturun
plot_bloch_multivector(statevector)
```

### Braket Gösterimi

Braket gösterimi, kuantum mekaniğinde kullandığımız matematiksel notaasyon sistemidir. Ket-ve-bra notaasyonu olarak da bilinir. Bu gösterim, kuantum durumlarını ve operatörleri ifade etmek için kullanılır. Gösterimi, ket |⟩ ve bra ⟨| sembolleri ile oluşturulur.

Ket gösterimi: |ψ⟩

Bu ifade, kuantum durum vektörünü temsil eder. Örneğin, |0⟩ ve |1⟩ qubit durumlarını temsil eder.
Bra gösterimi: ⟨ψ|

Bra gösterimi, bir vektörün eşleniğini temsil eder ve bir qubit durumunun kompleks konjugesi olarak düşünülebilir.
Bir iç çarpım ifadesi: ⟨ψ|φ⟩

Bu ifade, iki qubit durumu arasındaki iç çarpımı (olasılık amplitüdünün çarpımı) temsil eder.
Bra-ket gösterimi, kuantum mekaniğinde operatörlerin, durumların ve durum değişikliklerinin matematiksel temsilini sağlar. Özellikle, süperpozisyon ve dolanıklık gibi kuantum kavramlarının ifade edilmesinde kullanılır.






### Multi-qubit

Çok qubitli bir sistem, birden fazla qubitin bir araya gelerek oluşturduğu bir sistemdir. Örneğin, 2-qubitli bir sistemde 2 adet qubit bulunur. Genel olarak n-qubitli bir sistemde 2^n farklı durum kombinasyonu oluşur.

Örnek olarak, 2-qubitli bir sistemdeki durumlar:

|00⟩: İlk qubit "0", ikinci qubit "0" durumunda.
|01⟩: İlk qubit "0", ikinci qubit "1" durumunda.
|10⟩: İlk qubit "1", ikinci qubit "0" durumunda.
|11⟩: İlk qubit "1", ikinci qubit "1" durumunda.
Bu durumlar, bir 2-qubitli sistemin olası durumlarını gösterir. Durumlar, qubitlerin süperpozisyon ve dolanıklık özelliklerine göre birleşerek oluşur.



Bir n-qubitli sistemin durumu, karmaşık amplitüdlerle ifade edilir. Genel olarak |ψ⟩ şeklinde ifade edilir. Bir 2-qubitli sistemin durumu aşağıdaki gibi olabilir:

|ψ⟩ = α|00⟩ + β|01⟩ + γ|10⟩ + δ|11⟩



α, β, γ ve δ karmaşık sayılardır.
|00⟩, |01⟩, |10⟩ ve |11⟩ qubit durumlarını temsil eder.


Çok qubitli bir sistemde, qubitler bir araya gelerek karmaşık bir durum oluştururlar. Bu durum, süperpozisyon ve dolanıklık özelliklerini içerir. Her bir qubit, 0 ve 1 durumlarını aynı anda taşıyabilir ve bu durumlar birleşerek n-qubitli sistemde çeşitli durum kombinasyonları oluşturur. Her durum kombinasyonu, kuantum mekaniğindeki olasılık amplitüdleriyle belirtilir.

```python
# 3-qubitlik bir devre oluşturun
circuit = QuantumCircuit(3)

# Süperpozisyon durumu oluşturmak için Hadamard kapısı uygulayın
for qubit in range(3):
    circuit.h(qubit)

# İkinci qubite X kapısı uygulayın
circuit.x(1)

# 0. ve 2. qubitler arasında CNOT kapısı uygulayın
circuit.cx(0, 2)

# Devreyi simüle edin
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()

# Qubit durumunu yazdırın
print("Qubit Durumu:", statevector)

# Her bir qubit için Bloch küresi gösterimi oluşturun
for qubit in range(3):
    plot_bloch_multivector(statevector, title=f"{qubit}. Qubit Bloch Küresi")
```
